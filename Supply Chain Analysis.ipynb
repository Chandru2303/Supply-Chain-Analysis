## Key Performance Indicators (KPIs)

### Inventory & Demand KPIs


### Top-selling Products:

```python
top_selling = df.groupby('SKU')['Number of products sold'].sum().sort_values(ascending=False).head(10)
print("Top Selling Products:\n", top_selling)
```

### Revenue by Product Type

```python
revenue_by_type = df.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False)
print("Revenue by type: \n", revenue_by_type)
```

### Inventory Turnover

```python
df['Inventory Turnover'] = df['Number of products sold'] / df['Stock levels']
print(df[['SKU','Inventory Turnover']].sort_values(by = 'Inventory Turnover', ascending=False).head(10))
```

# Supplier & Manufacturing KPIs

### Average Lead Time per Supplier

```python
avg_lead_time_supplier = df.groupby('Supplier name')['Lead time'].mean()
print("Average Lead Time per Supplier:\n", avg_lead_time_supplier)
```

### Defect Rate per Supplier

```python
defect_rate = df.groupby('Supplier name')['Defect rates'].mean()
print("Defect Rates per Supplier:\n", defect_rate)
```

### Inspection Pass Rate

```python
inspection_pass_rate = df['Inspection results'].value_counts(normalize=True) * 100
print("Inspection Result Percentage:\n", inspection_pass_rate)
```

# Shipping & Logistics KPIs

### Average Shipping Time per Carrier

```python
shipping_time_carrier = df.groupby('Shipping carriers')['Shipping times'].mean()
print("Avg Shipping Time per Carrier:\n", shipping_time_carrier)
```

### Average Shipping Cost per Route 

```python
shipping_cost_route = df.groupby('Routes')['Shipping costs'].mean()
print("Avg Shipping Cost per Route:\n", shipping_cost_route)
```

### Transportation Cost by Mode

```python
transport_cost = df.groupby('Transportation modes')['Costs'].sum()
print("Transportation Cost by Mode:\n", transport_cost)
```

# Financial KPIs

### Total Revenue

```python
total_revenue = df['Revenue generated'].sum()
print("Total Revenue: $", total_revenue)
```

### Profit Estimation (basic)

```python
df['Estimated Profit'] = df['Revenue generated'] - (df['Manufacturing costs'] + df['Shipping costs'])
print("Total Estimated Profit: $", df['Estimated Profit'].sum())
```

## Export KPI Summary to CSV

```python
summary = {
    'Total Revenue': [total_revenue],
    'Total Estimated Profit': [df['Estimated Profit'].sum()],
    'Top Product SKU': [top_selling.index[0]],
    'Top Product Sales': [top_selling.values[0]]
}
summary_df = pd.DataFrame(summary)
summary_df.to_csv('kpi_summary.csv', index=False)
```

## Forecasting by SKU using Prophet

## Forecast for Top N SKUs or Product Types

## Loop and Forecast Each

```python
from prophet import Prophet
import pandas as pd
import os
import matplotlib.pyplot as plt

# Ensure forecast folder exists
os.makedirs("forecasts", exist_ok=True)

# Set parameters
group_field = "Product type"
forecast_target = "Revenue generated"  # or "Order quantities"

# Convert 'Order Date' to datetime and create week period
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Week'] = df['Order Date'].dt.to_period('W').dt.to_timestamp()

top_items = df[group_field].value_counts().head(5).index  # top 5 product types

for item in top_items:
    df_item = df[df[group_field] == item]
    df_grouped = df_item.groupby('Week')[forecast_target].sum().reset_index()
    df_grouped = df_grouped.rename(columns={'Week': 'ds', forecast_target: 'y'})

    if len(df_grouped) < 10:
        print(f"Skipping {item} due to insufficient data")
        continue

    model = Prophet()
    model.fit(df_grouped)

    future = model.make_future_dataframe(periods=6, freq='W')  # Forecasting for 6 weeks
    forecast = model.predict(future)

    # Save forecast as CSV
    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv(
        f"forecasts/{item}_{forecast_target.replace(' ', '_')}_forecast.csv", index=False
    )

    # Save chart as PNG
    fig = model.plot(forecast)
    plt.title(f"{item} - Forecast for {forecast_target}")
    fig.savefig(f"forecasts/{item}_{forecast_target.replace(' ', '_')}_forecast.png")

    print(f"Saved forecast for: {item}")
```
